```python
# Snake básico para Google Colab

tam = 10
x = 5
y = 5

seguir = "S"

while seguir.upper() == "S":

    # Dibujar tablero
    for fila in range(tam):
        for columna in range(tam):

            if fila == y and columna == x:
                print("O", end=" ")
            else:
                print(".", end=" ")

        print()

    tecla = input("Mover (W/A/S/D): ").upper()

    if tecla == "W":
        y -= 1

    elif tecla == "S":
        y += 1

    elif tecla == "A":
        x -= 1

    elif tecla == "D":
        x += 1

    # Perder al salir del tablero
    if x < 0 or x >= tam or y < 0 or y >= tam:

        print("¡Perdiste!")

        seguir = input("¿Desea continuar? (S/N): ").upper()

        if seguir == "S":
            x = 5
            y = 5

    print("\n" * 3)
    continuar = "S"

while continuar.upper() == "S":

    print("La serpiente se está moviendo...")

    perder = input("¿Perdiste? (S/N): ")

    if perder.upper() == "S":

        continuar = input("¿Desea continuar? (S/N): ")

print("Gracias por jugar")
```
